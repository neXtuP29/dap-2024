## Задание
1. Создать массив 10x10 со случайными значениями, найти минимум и максимум (низкая)

````python
import numpy as np

arr = np.random.random((10, 10))
print("Минимум:", arr.min())
print("Максимум:", arr.max())
````

2. Создать 8x8 матрицу и заполнить её единицами и нулями в шахматном порядке

````python
chessboard = np.zeros((8, 8), dtype=int)
chessboard[1::2, ::2] = 1
chessboard[::2, 1::2] = 1
print("Шахматная доска:")
print(chessboard)
````

3. Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8

````python
z = np.arange(11)
z[(z > 3) & (z < 8)] *= -1
print("Массив после изменения знака:")
print(z)
````

4. Найти среднее заначение строки матрицы и вычесть его из каждого елемента строки

````python
z = np.random.rand(5, 10)
row_means = z.mean(axis=1, keepdims=True)
z_normalized = z - row_means
print("Исходная матрица:")
print(z)
print("Нормализованная матрица:")
print(z_normalized)
````

5. Поменяйте 1 и 2 строки матрицы

````python
z = np.arange(25).reshape(5,5)
z[[0, 1]] = z[[1, 0]]
print("Матрица после замены строк:")
print(z)
````

6. Дан массив 10x2 подеставлющий собой точки в декартовой системе координат, преобразовать в полярную

````python
z = np.random.random((10,2))
x, y = z[:, 0], z[:, 1]
r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)
polar_coords = np.column_stack((r, theta))
print("Полярные координаты:")
print(polar_coords)
````

7. Заменить максимальный элемент массива на ноль

````python
z = np.random.random(10)
max_index = np.argmax(z)
z[max_index] = 0
print("Массив после замены максимума:")
print(z)
````

8. Дан массив, добавить 1 к каждому элементу с индексом, заданным в другом массиве. Добавление производиться только 1 раз

````python
z = np.ones(10)
i = np.random.randint(0, len(z), 20)
np.add.at(z, i, 1)
print("Массив после добавления:")
print(z)
````

9. Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением

````python
z = np.array([1, 2, 3, 4, 5])
result = np.zeros(len(z) + (len(z)-1)*3)
result[::4] = z
print("Вектор с нулями:")
print(result)
````

10. Найти ближайшее к заданному значению число в заданном массиве

````python
z = np.arange(100)
v = np.random.uniform(0, 100)
nearest = z[np.argmin(np.abs(z - v))]
print(f"Ближайшее к {v:.2f} число: {nearest}")
````

# Ответы на вопросы
## 1.Типы массивов:

```
# np.array([1, 4, 2, 5, 3]) → int64

np.array([1., 4, 2, 5, 3]) → float64

np.array([1, 4, 2, 5, 3], dtype='float128') → float128

np.array([1., 4, 2, 5, 3], dtype='complex128') → complex128

np.array([1, 4, 2, 5, 3], dtype='float64') → float64

np.array([1, 4, 2, 5, 3], dtype=int64) → int64

np.array([1, 4, 2, 5, 3], dtype='bool') → bool

np.array([1, 4, 2, 5, 3], dtype='str') → str

np.array([1, 4, 2, 5, '3']) → str

np.array([1, -4, 2, 0, 1j]) → complex128
```

## 2.Результаты операций:

```
# np.array([1, 4, 2, 5, 3], dtype='str')[1] + '1' - '41' (конкатенация строк)
np.array([1, -4, 2, 0, 3], dtype='bool')[3] or np.array([1, -4, 2, 0, 3])[1] - -4 (0 or -4 = -4)
np.array([1, -4, 2, 0, 3], dtype='bool')[3] and np.array([1, -4, 2, 0, 3])[1] - 0 (0 and -4 = 0)
np.array([1, -4, 2, 0, True], dtype='int64')[-1] - 1 (True преобразуется в 1)
np.array([1, -4, 2, 0, 1j], dtype='int64')[1] - Ошибка (нельзя преобразовать комплексное число в int)
np.array([1, -4, 2, 0, 1j], dtype='str')[-1] - '1j'
np.array([1, 4, 2, 5, '3'])[-1] * 3 -'333'
```

## 3.Результаты создания массивов:

```
np.ones((3, 5), dtype=float) - Массив 3x5 из единиц типа float
np.full((3, 5), 3.14) - Массив 3x5 со значением 3.14
np.linspace(0, 1, 5) - [0. 0.25 0.5 0.75 1. ]
np.diag((1,3,5)) - Диагональная матрица с элементами 1,3,5
np.smile((1,3,5)) - Ошибка (нет функции np.smile)
np.random.random((3, 3)) - Случайная матрица 3x3
np.eye(3) - Единичная матрица 3x3
np.empty(3) - Неинициализированный массив размера 3
```

