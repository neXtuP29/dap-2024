6.Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
young_survived = df[(df['Age'] < 30) & (df['Survived'] == 1)].shape[0]
young_total = df[df['Age'] < 30].shape[0]
young_survival_rate = young_survived / young_total if young_total > 0 else 0

old_survived = df[(df['Age'] > 60) & (df['Survived'] == 1)].shape[0]
old_total = df[df['Age'] > 60].shape[0]
old_survival_rate = old_survived / old_total if old_total > 0 else 0

print(f"Доля выживших младше 30 лет: {young_survival_rate:.2f} ({young_survived}/{young_total})")
print(f"Доля выживших старше 60 лет: {old_survival_rate:.2f} ({old_survived}/{old_total})")
print(f"Люди моложе 30 лет выживали чаще: {young_survival_rate > old_survival_rate}")
print()
```

7.Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

```python
women_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)].shape[0]
women_total = df[df['Sex'] == 'female'].shape[0]
women_survival_rate = women_survived / women_total

men_survived = df[(df['Sex'] == 'male') & (df['Survived'] == 1)].shape[0]
men_total = df[df['Sex'] == 'male'].shape[0]
men_survival_rate = men_survived / men_total

print(f"Доля выживших женщин: {women_survival_rate:.2f} ({women_survived}/{women_total})")
print(f"Доля выживших мужчин: {men_survival_rate:.2f} ({men_survived}/{men_total})")
print(f"Женщины выживали чаще мужчин: {women_survival_rate > men_survival_rate}")
print()
```

8.Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
def extract_first_name(full_name):
    """Извлекает первое имя из полного имени"""
    if 'Mr.' in full_name:
        parts = full_name.split('Mr. ')
        if len(parts) > 1:
            name_part = parts[1].split(' ')[0]
            # Убираем кавычки и точки
            name_part = name_part.replace('"', '').replace('.', '')
            return name_part
    elif 'Master.' in full_name:
        parts = full_name.split('Master. ')
        if len(parts) > 1:
            name_part = parts[1].split(' ')[0]
            name_part = name_part.replace('"', '').replace('.', '')
            return name_part
    return None

male_passengers = df[df['Sex'] == 'male']
male_names = male_passengers['Name'].apply(extract_first_name).dropna()
most_common_name = male_names.mode()[0] if not male_names.empty else "Не найдено"
print(f"Самое популярное имя среди мужчин: {most_common_name}")
print(f"Количество мужчин с этим именем: {male_names.value_counts().get(most_common_name, 0)}")
print()
```

9.Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания? Выберите верные утверждения:

В среднем мужчины 1-го класса старше 40 лет В среднем женщины 1-го класса старше 40 лет Мужчины всех классов в среднем старше женщин того же класса В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса

```python
age_by_class_sex = df.groupby(['Pclass', 'Sex'])['Age'].mean()
print("Средний возраст по классам и полу:")
for (pclass, sex), mean_age in age_by_class_sex.items():
    print(f"Класс {pclass}, {sex}: {mean_age:.1f} лет")

# Проверка утверждений
men_first_class_age = age_by_class_sex.get((1, 'male'), 0)
women_first_class_age = age_by_class_sex.get((1, 'female'), 0)

print(f"\nПроверка утверждений:")
print(f"Мужчины 1-го класса старше 40 лет: {men_first_class_age > 40}")
print(f"Женщины 1-го класса старше 40 лет: {women_first_class_age > 40}")

# Сравнение среднего возраста мужчин и женщин по классам
men_women_comparison = []
for pclass in [1, 2, 3]:
    men_age = age_by_class_sex.get((pclass, 'male'), 0)
    women_age = age_by_class_sex.get((pclass, 'female'), 0)
    men_women_comparison.append(men_age > women_age)

print(f"Мужчины всех классов старше женщин: {all(men_women_comparison)}")

# Сравнение среднего возраста по классам
class_ages = df.groupby('Pclass')['Age'].mean()
age_decreasing = class_ages[1] > class_ages[2] > class_ages[3]
print(f"Люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса: {age_decreasing}")
print()
```
